Data Manipulating:

```python
import pandas as pd
import numpy as np

df = pd.read_csv('gym_members_exercise_tracking.csv')
threshold = 100
filtered_df = df[df['Avg_BPM'] > threshold]

# Group and aggregate
grouped = df.groupby("Age").agg({'Fat_Percentage': 'mean'})

df['Good'] = df['Resting_BPM'] * 2

df.to_csv('new_dataset.csv', index=False)
```
